```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("Titanic-Dataset.csv")
```

```python
print("\nSummary Statistics (Numerical):")
data.describe()
```

```python
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']

# Histograms
plt.figure(figsize=(14, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data[col].dropna(), kde=True, bins=30, color='skyblue')
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
# Boxplots
plt.figure(figsize=(14, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x=data[col], color='salmon')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
corr = data[numeric_cols + ['Survived', 'Pclass']].corr()
corr
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(data[numeric_cols + ['Survived']], hue='Survived', palette='Set1', diag_kind='kde')
plt.suptitle("Pairplot of Numeric Features Colored by Survival", y=1.02)
plt.show()
```
